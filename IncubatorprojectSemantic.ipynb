```python
#To download the scrapped data from the webpage as a csv file for the title Tracking UK Legislation
import requests
from bs4 import BeautifulSoup
import csv

# URL of the page
url = "https://www.legislation.gov.uk/new/all/2025-05-20"

try:
    # Fetch the web page
    response = requests.get(url)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
    soup = BeautifulSoup(response.content, "html.parser")

    # F the div containing the list
    # Based on your snippet, the div has class="content"
    content_div = soup.find('div', class_='content')

    if content_div:
        # Found all list items within the div
        list_items = content_div.find_all('li')

        if list_items:
            # Initialized a list to hold the extracted text
            extracted_data = []

            # Extracted the text from each list item
            for item in list_items:
                # Get the text, stripping leading/trailing whitespace
                item_text = item.get_text(strip=True)
                extracted_data.append([item_text]) # Append as a list for CSV writer

            # Saved the data to a CSV file
            csv_filename = "legislation_list_data.csv"
            with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
                csv_writer = csv.writer(csvfile)
                # csv_writer.writerow(['Legislation Info'])
                csv_writer.writerows(extracted_data)

            print(f"List data successfully extracted and saved to {csv_filename}")

        else:
            print("No list items (<li>) found within the content div.")
    else:
        print("Content div with class 'content' not found on the page.")

except requests.exceptions.RequestException as e:
    print(f"Error fetching the page: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# etl/pipeline.py

import os
from etl.downloader import download_legislation
from etl.cleaner import clean_html
from etl.extractor import extract_metadata
from etl.embedder import embed_texts
from db.sql_connector import store_metadata
from vector_store.vector_connector import store_embeddings


def main():
    # Load environment variables
    year = os.getenv("LEGISLATION_YEAR", "2024")
    month = os.getenv("LEGISLATION_MONTH", "08")
    category = os.getenv("LEGISLATION_CATEGORY", "planning")

    print(f"Starting ETL for {category} ({month}/{year})")

    # Step 1: Download data from legislation.gov.uk
    raw_documents = download_legislation(year, month, category)

    # Step 2: Clean and extract metadata
    cleaned_documents = []
    metadata_list = []
    for html_doc in raw_documents:
        cleaned_text = clean_html(html_doc['html'])
        metadata = extract_metadata(html_doc['html'])
        metadata['content'] = cleaned_text
        cleaned_documents.append(cleaned_text)
        metadata_list.append(metadata)

    # Step 3: Store metadata and content in SQL
    ids = store_metadata(metadata_list)

    # Step 4: Generate embeddings using sentence-transformers/all-MiniLM-L6-v2
    embeddings = embed_texts(cleaned_documents)

    # Step 5: Store embeddings in vector database
    store_embeddings(ids, embeddings)

    print("ETL process completed successfully.")


if __name__ == "__main__":
    main()
```

```python
# etl/downloader.py

import requests
from bs4 import BeautifulSoup
import time

BASE_URL = "https://www.legislation.gov.uk/new/all/2025-05-20"
HEADERS = {"User-Agent": "Mozilla/5.0"}

# Get parameters from user input
year = input("Enter the legislation year (e.g., 2025): ")
month = input("Enter the legislation month (e.g., 05): ")
category = input("Enter the legislation category (e.g., planning): ")


print("Downloading legislation documents...")
results = []

# The actual site doesn't provide an exact API, so we use a search URL
search_url = f"{BASE_URL}/all?title={category}&year={year}"
print(f"Searching: {search_url}")

response = requests.get(search_url, headers=HEADERS)
soup = BeautifulSoup(response.content, "html.parser")

links = soup.select(".result-title a")
print(f"Found {len(links)} links")

for link in links[:10]:  # limit to 10 for testing
    href = link.get("href")
    full_url = BASE_URL + href

    try:
        doc_resp = requests.get(full_url, headers=HEADERS)
        doc_soup = BeautifulSoup(doc_resp.content, "html.parser")

        # Get the HTML content inside the legislation text container
        content_div = doc_soup.select_one("#content")
        if not content_div:
            continue

        html_content = str(content_div)
        results.append({"url": full_url, "html": html_content})
        print(f"Downloaded: {full_url}")

        time.sleep(1)  # be polite to the server
    except Exception as e:
        print(f"Failed to download {full_url}: {e}")

# 'results' variable now contains the downloaded data
# You can process or save 'results' here
print(f"Finished downloading. Collected {len(results)} documents.")
```
